```python
def BoxPlot(değişken):
    plt.figure()
    plt.boxplot(x=değişken,notch=True,flierprops={"markerfacecolor":"r"},data=df)
    plt.xlabel(değişken)
    plt.ylabel("size")
    plt.title("Boxplot {}".format(değişken))
    plt.show()
```

```python
for değişken in list(df.columns):
    if df[değişken].dtype == "int64":
        BoxPlot(değişken)
    elif df[değişken].dtype == "float64":
        BoxPlot(değişken)
    else:
        pass
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=50,contamination=0.5)
```

```python
lof_pred = lof.fit_predict(df.drop("DateTime",axis=1))
```

```python
df_scores = lof.negative_outlier_factor_
```

```python
scores = pd.DataFrame(np.sort(df_scores))
```

```python
scores.plot(stacked=True,xlim=[0,100],style=".-");
```

```python
eşik_değer = np.sort(df_scores)[17]
```

```python
df = df[df_scores>eşik_değer]
```

```python
df = pd.DataFrame(df.to_records(index=False))
```

```python
df.info()
```

```python
def HistogramPlot(değişken):
     
    plt.figure()
    plt.hist(df[değişken], bins = 85, color = "orange")
    plt.xlabel(değişken)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(değişken))
    plt.show()
```

```python
değişkenler = df.drop("DateTime",axis=1)
for değişken in değişkenler:
    HistogramPlot(değişken)
```
